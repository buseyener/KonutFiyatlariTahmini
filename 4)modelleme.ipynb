```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
```

```python
# Tahminleri ve gerçek değerleri alma
tahmin = reg.predict([[17, 94, 3, 7, 5, 2, 1, 140, 5, 5,0,1]])
gercek = df["ilan_fiyat"]

# Hata paylarını hesaplama
hatalar = gercek - tahmin

# Hata paylarının grafiği
plt.figure(figsize=(10, 6))
plt.scatter(np.arange(len(hatalar)), hatalar, color='blue', label='Hata Payları')
plt.axhline(y=0, color='red', linestyle='--', label='Sıfır Hata')
plt.title('Hata Payları Grafiği')
plt.xlabel('Veri Noktası')
plt.ylabel('Hata')
plt.legend()
plt.grid(True)
plt.show()
```

```python
skor = reg.score(df[["ilan_ilce",  "net_metrekare", "oda_sayısı", "bulunduğu_kat", "ısıtma_tipi", "banyo_sayısı", "tipi",
         "brüt_metrekare", "binanın_yaşı", "binanın_kat_sayısı","site_içerisinde","krediye_uygunluk"]],df["ilan_fiyat"])
print("Model skoru:", skor)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Tahminleri alma
tahminler = reg.predict(df[["ilan_ilce",  "net_metrekare", "oda_sayısı", "bulunduğu_kat", "ısıtma_tipi", "banyo_sayısı", "tipi",
         "brüt_metrekare", "binanın_yaşı", "binanın_kat_sayısı","site_içerisinde","krediye_uygunluk"]])
```

```python
# Ortalama Mutlak Hata (Mean Absolute Error - MAE)
mae = mean_absolute_error(df["ilan_fiyat"], tahminler)
print("Ortalama Mutlak Hata (MAE):", mae)
```

```python
# Ortalama Kare Hatası (Mean Squared Error - MSE)
mse = mean_squared_error(df["ilan_fiyat"], tahminler)
print("Ortalama Kare Hatası (MSE):", mse)
```

```python
# Ortalama Mutlak Yüzde Hata (Mean Absolute Percentage Error - MAPE)
mape = np.mean(np.abs((df["ilan_fiyat"] - tahminler) / df["ilan_fiyat"])) * 100
print("Ortalama Mutlak Yüzde Hata (MAPE):", mape)
```

```python
# R-kare (R-squared)
r_squared = reg.score(df[["ilan_ilce",  "net_metrekare", "oda_sayısı", "bulunduğu_kat", "ısıtma_tipi", "banyo_sayısı", "tipi",
         "brüt_metrekare", "binanın_yaşı", "binanın_kat_sayısı","site_içerisinde","krediye_uygunluk"]],df["ilan_fiyat"])
print("R-kare (R-squared):", r_squared)
```
